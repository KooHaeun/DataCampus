```python
info1 = list()

with mp_hands.Hands(min_detection_confidence = 0.80, min_tracking_confidence = 0.5, max_num_hands=1) as hands:
  # confidence 조절해서 해야함
      for i in range(0,len(listdir(path))):  # len(listdir(path))
          cap = cv2.VideoCapture(path+listdir(path)[i])
          #while cap.isOpened(): ##비디오에 사용?
          ret, frame = cap.read()

          # BGR 2 RGB
          image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

          # Flip on horizontal
          # image = cv2.flip(image, 1)

          # Set flag
          image.flags.writeable = False

          # Detections
          results = hands.process(image)

          # Set flag to true
          image.flags.writeable = True

          # RGB 2 BGR
          image = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

          # Detections
          print('{} {}'.format(results,i))        
          
          if results.multi_hand_landmarks:
            landmarks = results.multi_hand_landmarks    
            imagename = listdir(path)[i]
            # landmarks = 
            info1.append((landmarks,imagename))
          if i%25 == 0:
            time.sleep(3)


          if cv2.waitKey(10) & 0xFF == ord('q') :
              break
        
cap.release()
cv2.destroyAllWindows()
```
