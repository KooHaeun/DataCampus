# 이제부터 크기 정규화 해야함.

그러기 위해서는 벡터를 벡터의 크기 데이터로 뽑아내야한다.




2. 크기 정규화는 
기존의 벡터의 크기에서 min 값과 max 값의 차이가가장 작은 것을 기준으로 할지,\
혹은 min 값이나 max 값이 가장 작은 것으로 할지,\
혹은 예전에 이야기 했던 2,3번 랜드마크로 할지

다 해봐야한다

```python
df_vector.describe()
```

```python
df_size = df_vector.copy()
```

```python
column_size =[]

for i in range(0,16):
  column_size.append('size_{1}-{0}'.format(i,i+1))
```

```python
column_size
```

```python
for i in range(0,16):
  df_size['x_{0}-{1}'.format(i+1,i)] = (df_vector['x_{0}-{1}'.format(i+1,i)]**2+df_vector['y_{0}-{1}'.format(i+1,i)]**2)**0.5
  print('{0}번째 x 작업'.format(i+1))
  # print(i+1), print(i)

  # 벡터의 크기 데이터를 만드는 코드
```

```python
df_size
```

```python
for i in range(0,16):
  df_size = df_size.drop('y_{0}-{1}'.format(i+1,i), axis=1)


df_size = df_size.drop(['y', 'image'], axis=1)
# 기존의 데이터에서 크기와 관련이 없어 필요 없는 데이터 삭제
```

```python
df_size.columns = column_size
df_size

# 벡터의 크기 데이터 완성
```

```python
df_size.describe()
```

```python
# 벡터 크기의 min 값과 max 의 차이가 가장 작은 벡터를 찾는 코드

minmax = []
for i in range(0,16):
  minmax.append(max(df_size['size_{1}-{0}'.format(i, i+1)])-min(df_size['size_{1}-{0}'.format(i, i+1)]))

print(min(minmax))
print(minmax)
minmax.index(min(minmax))
```

15로 나왔으니 결국 16-15 데이터를 쓰면 된다.

모든 벡터 다 통틀어서 벡터의 max값과 min 값의 차이가 가장 작은것, 즉, 데이터간 이 벡터만큼은 크게 변화가 없는 것. 그것이 16-15 벡터가 될 수 있다는 말.

그래서 16-15 벡터의 크기를 1로 단위화 시키고 그만큼 다른 벡터들의 크기를 보정해서 벡터에 반영해야함.
