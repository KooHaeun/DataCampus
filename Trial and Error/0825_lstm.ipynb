```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
X=tf.reshape(np.array(train.iloc[:,:-2]),[-1, 32, 35]) # 마지막은 열 개수
X.shape

#39,32,35
```

```python
y = tf.keras.utils.to_categorical(train_labels['label']) 
y.shape

#39,2
```

```python
model = Sequential()
model.add(LSTM(32, input_shape=(32,35))) #개수, 열
model.add(Dense(64, activation='relu'))
model.add(Dense(2, activation='softmax')) #label 개수

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X,y, epochs=20, batch_size=128, validation_split=0.2)
```

```python
y #왼쪽 올바르지 않은, 오른쪽 올바른
```

```python
test_X=tf.reshape(np.array(test.iloc[:,:-2]),[-1, 32, 35]) # 마지막은 열 개수
test_X.shape
```

```python
prediction = model.predict(test_X)
```
